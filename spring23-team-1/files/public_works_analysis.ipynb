```python
def extract_first_part(neighborhood):
    return neighborhood.split("/")[0]

def extract_second_part(neighborhood):
    return neighborhood.split("/")[1]
```

```python
# Extract the neighborhood parts for rows that contain '/'
for i in range(len(public_works_df)):
    if '/' in public_works_df.loc[i, 'Neighborhood']:
        public_works_df.loc[i, 'Neighborhood_1'] = extract_first_part(public_works_df.loc[i, 'Neighborhood'])
        public_works_df.loc[i, 'Neighborhood_2'] = extract_second_part(public_works_df.loc[i, 'Neighborhood'])
```

```python
from geopy.geocoders import Nominatim

# define a function to extract the latitude and longitude for an address
def get_lat_long(row):
    print(row['final_address'])
    try:
        geolocator = Nominatim(user_agent='test-gunner')
        location = geolocator.geocode(row['final_address'], timeout=10)
        if location is None:
            location = geolocator.geocode(row['final_address_1'], timeout=10)
            if location is None:
                location = geolocator.geocode(row['final_address_2'], timeout=10)
                if location is None:
                    return pd.Series([None, None])
        return pd.Series([location.latitude, location.longitude])
    except Exception as e:
        print('Error occurred while geocoding address {}: {}'.format(row['final_address'], e))
        return pd.Series([None, None])

# apply the function to the dataframe and store the latitude and longitude values in new columns
lat_long_df = public_works_df.apply(get_lat_long, axis=1)

# check if the length of the lat_long_df is equal to the length of the public_works_df
if len(lat_long_df) != len(public_works_df):
    print('Error: length of lat_long_df ({}) is not equal to length of public_works_df ({})'.format(len(lat_long_df), len(public_works_df)))
else:
    # assign the latitude and longitude values to the public_works_df dataframe
    public_works_df[['latitude', 'longitude']] = lat_long_df

print('completed')
```

```python
from geopy.geocoders import Nominatim
from geopy.point import Point
import numpy as np

geolocator = Nominatim(user_agent = "test")

def reverse_geocoding(latitude, longitude):
    try:
        location = geolocator.reverse(Point(latitude, longitude))
        return location.raw['display_name']
    except:
        return None
```

```python
import folium
import pandas as pd

#create a map
Redzones = folium.Map(prefer_canvas = True)

def plotDot(point):
    '''input: series that contains a numeric named latitude and a numeric named longitude
    this function creates a CircleMarker and adds it to your this_map'''

    popup_text = 'Red Zones'

    folium.CircleMarker(location = [point.latitude, point.longitude],
                        popup = popup_text,
                        radius = 5,
                        weight = 5,
                        color = 'red').add_to(Redzones)

# use df.apply(,axis=1) to "iterate" through every row in your dataframe

# temp = ramps_0_df.head(100)

red_zones = public_works_df[(public_works_df['Sidewalk_Plates_In_Use'].isna() == False)]
red_zones.apply(plotDot, axis = 1)

#Set the zoom to the maximum possible
Redzones.fit_bounds(Redzones.get_bounds())

Redzones.save("../maps/public-works/redzones.html")

Redzones
```
