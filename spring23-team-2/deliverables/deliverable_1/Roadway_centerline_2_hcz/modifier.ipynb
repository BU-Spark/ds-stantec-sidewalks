```python
import pandas as pd
import matplotlib.pyplot as plt



# Read the CSV file
data = pd.read_csv('Roadway_Centerline_2.csv')

# Create a histogram of road lengths for each jurisdiction
for jurisdic in data['JURISDICTI'].unique():
    plt.hist(data[data['JURISDICTI'] == jurisdic]['LENGTH'], alpha=0.5, label=jurisdic)

# Set plot title and axis labels
plt.title('Distribution of Road Lengths by Jurisdiction')
plt.xlabel('Road Length (ft)')
plt.ylabel('Frequency')

# Add a legend to the plot
plt.legend()

# Display the plot
plt.show()
```
