### Ramps_with_score is the sorted data with the ramps scores.

```python
import pandas as pd

df = pd.read_csv('data/Ramps_with_score.csv')
```

```python
import matplotlib.pyplot as plt
scores = df["score_1"]
scores.head()
bins= [0,20,40,60,80,101]
segments = pd.cut(scores,bins,right = False)
counts = pd.value_counts(segments,sort=False)

plt.title("The distribution of Accessibility scores of Ramps")
b = plt.bar(counts.index.astype(str),counts)
plt.show()

bins2= [0,5,10,15,21]
segments2 = pd.cut(scores,bins2,right = False)
counts2 = pd.value_counts(segments2,sort=False)

plt.title("The distribution of Accessibility scores of Ramps")
b = plt.bar(counts2.index.astype(str),counts2)
plt.show()
```
